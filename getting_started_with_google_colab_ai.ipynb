ANALYSE ET EXPLORATION DE DONNEES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from pathlib import Path
from google.colab import files

print("Bibliothèques chargées avec succès.")
```

```python
# style global des graphiques
sns.set_style("whitegrid")

plt.rcParams["figure.figsize"] = (12, 6)
plt.rcParams["axes.titlesize"] = 14
plt.rcParams["axes.labelsize"] = 12

print("Style configuré.")
```

CHARGEMENT DES DONNÉES

```python
print("Téléversez votre fichier Coffe_sales.csv")

uploaded = files.upload()

file_path = list(uploaded.keys())[0]

df = pd.read_csv(file_path)

print("Fichier chargé :", file_path)
```

```python
#APERÇU DU DATASET
print("Dimensions :", df.shape)

display(df.head())
display(df.tail())
```

EXPLORATION INITIALE

```python
#INFORMATIONS GÉNÉRALES
print("\n--- INFORMATIONS GÉNÉRALES ---")
df.info()

print("\n--- NOMS DES COLONNES ---")
print(df.columns.tolist())

print("\n--- VALEURS MANQUANTES ---")
missing = df.isna().sum().sort_values(ascending=False)
display(missing[missing > 0])

print("\n--- DOUBLONS ---")
print("Nombre de lignes dupliquées :", df.duplicated().sum())

print("\n--- STATISTIQUES NUMÉRIQUES ---")
display(df.describe(include=[np.number]).T)

print("\n--- STATISTIQUES CATÉGORIELLES ---")
display(df.describe(include=["object"]).T)
```

NETTOYAGE DES COLONNES

```python
#Nettoyage Léger des noms de colonnes
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
# Harmonisation si certaines colonnes existent sous des noms proches
rename_map = {
    "timeofday": "time_of_day",
    "monthname": "month_name",
    "weekdaysort": "weekday_sort",
    "monthsort": "month_sort",
    "hourofday": "hour_of_day",
    "coffeename": "coffee_name",
    "cashtype": "cash_type"
}

df.rename(columns=rename_map, inplace=True)

print(df.columns.tolist())
```

```python
# CONVERSION DES TYPES
numeric_cols = [
    "hour_of_day",
    "money",
    "weekday_sort",
    "month_sort"
]

for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")

# Nettoyage texte
for col in df.select_dtypes(include="object").columns:
    df[col] = df[col].astype(str).str.strip()

print("Nettoyage terminé.")
```

COPIE PROPRE DES DONNÉES

```python
df_clean = df.copy()

if "coffee_name" in df_clean.columns:
    df_clean["coffee_name"] = (
        df_clean["coffee_name"]
        .str.lower()
        .str.strip()
    )

print("Dataset nettoyé.")
```

RÉPARTITION & RÉSUMÉ DES VENTES

```python
print("\n--- VARIABLES CATÉGORIELLES ---")

for col in [
    "cash_type",
    "coffee_name",
    "time_of_day",
    "weekday",
    "month_name"
]:
    if col in df_clean.columns:
        print(f"\n{col}")
        display(df_clean[col].value_counts().head(10))

print("\n--- RÉSUMÉ DES VENTES ---")

if "money" in df_clean.columns:
    print("Total :", round(df_clean["money"].sum(), 2))
    print("Moyenne :", round(df_clean["money"].mean(), 2))
    print("Médiane :", round(df_clean["money"].median(), 2))
    print("Min :", round(df_clean["money"].min(), 2))
    print("Max :", round(df_clean["money"].max(), 2))
```

ANALYSES DESCRIPTIVE

```python
results = {}

if {"coffee_name", "money"}.issubset(df_clean.columns):
    results["sales_by_coffee"] = (
        df_clean.groupby("coffee_name")["money"]
        .agg(["count", "sum", "mean"])
        .sort_values("sum", ascending=False)
    )

if {"weekday", "money"}.issubset(df_clean.columns):
    results["sales_by_weekday"] = (
        df_clean.groupby("weekday")["money"]
        .agg(["count", "sum", "mean"])
        .sort_values("sum", ascending=False)
    )

if {"hour_of_day", "money"}.issubset(df_clean.columns):
    results["sales_by_hour"] = (
        df_clean.groupby("hour_of_day")["money"]
        .agg(["count", "sum", "mean"])
        .sort_index()
    )

for name, table in results.items():
    print(f"\n--- {name} ---")
    display(table.head(10))
```

TOP 10 BOISSONS

```python
if {"coffee_name", "money"}.issubset(df_clean.columns):

    top_coffees = (
        df_clean.groupby("coffee_name")["money"]
        .sum()
        .sort_values(ascending=False)
        .head(10)
    )

    plt.figure(figsize=(12, 6))
    plt.barh(top_coffees.index, top_coffees.values)
    plt.title("Top 10 boissons par chiffre d'affaires")
    plt.xlabel("Chiffre d'affaires")
    plt.gca().invert_yaxis()
    plt.show()
```

VENTES PAR HEURE (CORRIGÉ)

```python
if {"hour_of_day", "money"}.issubset(df_clean.columns):

    hourly_sales = (
        df_clean.groupby("hour_of_day")["money"]
        .sum()
    )

    plt.figure(figsize=(12, 6))
    plt.plot(
        hourly_sales.index,
        hourly_sales.values,
        marker="o"
    )

    plt.title("Chiffre d'affaires par heure")
    plt.xlabel("Heure")
    plt.ylabel("Montant")

    plt.grid(True)
    plt.show()
```

HISTOGRAMME (CORRIGÉ)

```python
if "money" in df_clean.columns:

    plt.figure(figsize=(10, 5))

    plt.hist(
        df_clean["money"].dropna(),
        bins=30
    )

    plt.title("Distribution des montants")
    plt.xlabel("Montant")
    plt.ylabel("Fréquence")

    plt.show()
```

MATRICE DE CORRÉLATION

```python
numeric_df = df_clean.select_dtypes(include=np.number)

if not numeric_df.empty:

    plt.figure(figsize=(10, 6))

    corr = numeric_df.corr()

    plt.imshow(corr)
    plt.colorbar()

    plt.xticks(
        range(len(corr.columns)),
        corr.columns,
        rotation=90
    )

    plt.yticks(
        range(len(corr.columns)),
        corr.columns
    )

    plt.title("Matrice de corrélation")

    plt.show()
```

EXPORT DES FICHIERS

```python
output_dir = Path("output")
output_dir.mkdir(exist_ok=True)

df_clean.to_csv(
    output_dir / "coffee_sales_cleaned.csv",
    index=False
)

for name, table in results.items():
    table.to_csv(
        output_dir / f"{name}.csv"
    )

print("Analyse terminée.")
print("Fichiers exportés dans output/")
```
