```python

#a. les operation de base

#1.Addition
def addition(a, b):
    return a + b

#2.Soustraction
def soustraction(a, b):
    return a - b

#3.Multiplication
def multiplication(a, b):
    return a * b

#4.Division
def division(a, b):
    if b != 0:
        return a / b
    else:
        return "Division impossible par zéro"
#Exemple

print(addition(10, 2))        # 12
print(soustraction(8, 2))    # 4
print(multiplication(6, 2))  # 12
print(division(8, 2))        # 4.0

#b. Calculs entre vecteurs

#1.Addition de vecteurs
def add_vecteurs(v1, v2):
    return [v1[i] + v2[i] for i in range(len(v1))]

#2.Produit scalaire
def produit_scalaire(v1, v2):
    return sum(v1[i] * v2[i] for i in range(len(v1)))

#Exemple

v1 = [1, 2, 3]
v2 = [4, 5, 6]

print(add_vecteurs(v1, v2))
print(produit_scalaire(v1, v2))
#Résultat :

[5, 7, 9]
32

#c. Calculs entre matrices

#1.Addition de matrices

def add_matrices(A, B):
    return [[A[i][j] + B[i][j] for j in range(len(A[0]))]
            for i in range(len(A))]


#2.Multiplication de matrices

def mult_matrices(A, B):
    resultat = [[0 for j in range(len(B[0]))]
                for i in range(len(A))]

    for i in range(len(A)):
        for j in range(len(B[0])):
            for k in range(len(B)):
                resultat[i][j] += A[i][k] * B[k][j]

    return resultat

#Exemple

A = [[1, 2],
     [3, 4]]

B = [[5, 6],
     [7, 8]]

print(add_matrices(A, B))
print(mult_matrices(A, B))
#Résultat :

[[6, 8], [10, 12]]

[[19, 22],
 [43, 50]]
#d. Avec NumPy (plus simple)

import numpy as np

A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

print(A + B)
print(A @ B)
#+ → addition
#@ → multiplication matricielle
```
